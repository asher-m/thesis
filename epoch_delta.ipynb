```python
import matplotlib.pyplot as plt

# get *correct* data module for joyce-apj-tab2
import meeting_20210302_data as data
```

```python
d = data.Data('joyce-apj-tab2')
d.read_data()
```

```python
d.datasets
```

```python
plt.figure(figsize=(16, 9))

for i in range(0, 5):
    plt.plot((d.eventdata[i]['psp_isois-epilo_l2-ic']['Epoch_ChanP'] - d.eventdata[i]['psp_isois-epilo_l2-ic']['Epoch_ChanP'][0]) / 1e9,
             d.eventdata[i]['psp_isois-epilo_l2-ic']['Epoch_ChanP_DELTA'] / 1e9,
             label=f'event {i}')

plt.xlabel('seconds from start')
plt.ylabel('delta seconds')
plt.legend()
```
