```python
import matplotlib.pyplot as plt

# get *correct* data module for joyce-apj-tab2
import data_jat as data
```

```python
d = data.Data('joyce-apj-tab2')
d.read_data()
```

```python
d.datasets
```

```python
plt.figure(figsize=(16, 9))

for i in range(0, 5):
    plt.plot((d.eventdata[i]['ChanP']['epoch'] - d.eventdata[i]['ChanP']['epoch'][0]) / 1e9,
             d.eventdata[i]['ChanP']['epoch_delta'] / 1e9,
             label=f'event {i}')

plt.xlabel('seconds from start')
plt.ylabel('delta seconds')
plt.legend()
```
